## 获取目前标签类型和数量

```python
import json
import os
import numpy as np

cwd = os.getcwd()

# 从字典里面找到key对应的value
def get_value(dictionary, key_find):
    val_find = ''
    for key, val in dictionary.items():
        if key == key_find:
            val_find = val
    return val_find

def save_json(path, file_name, content):
    if not os.path.exists(path):
        os.makedirs(path)

    json_str = json.dumps(content, ensure_ascii=False, indent=4)
    with open(f'{path}/{file_name}', 'a', encoding='utf-8')as json_file:
        json_file.write(json_str)


# 读取json文件内容,返回字典格式
def get_all_labels(attr):
    if attr == 'factory':
        attr_name = 'MARKERS_RESULTS_AS_MANUFACTURER'
    elif attr == 'type':
        attr_name = 'MARKERS_RESULTS_AS_TYPE'
    with open(f'./labels/check_result_c2.json','r',encoding='utf-8')as fp:
        json_data = json.load(fp)
        # print('这是文件中的json数据：',json_data)
        # print('这是读取到文件数据的数据类型：', type(json_data))
    factories = []
    for key, contend in json_data.items():
        factory_get = get_value(contend, attr_name)
        #if factory_get != '':
        factories.append(factory_get)
    b = np.unique(factories)
    factories_json = {}
    for item in b:
        factories_json[item] = factories.count(item)
    factories_json = dict(sorted(factories_json.items(),key=lambda x:x[1],reverse=True))
    print(factories_json, len(factories_json))
    print('data_len:',len(json_data))
    # save_json('./labels',f'{attr}.json',factories_json)
        

get_all_labels('type')
# get_all_labels('type')
# 1 factory 0 type
        
```

## 剔除不需要的标签

```python
# 读取json文件内容,返回字典格式
def pick_labels(attr, baseline, open_file, save_file):
    if attr == 'factory':
        attr_name = 'MARKERS_RESULTS_AS_MANUFACTURER'
    elif attr == 'type':
        attr_name = 'MARKERS_RESULTS_AS_TYPE'
    with open(open_file,'r',encoding='utf8')as fp:
        json_data = json.load(fp)
        # print('这是文件中的json数据：',json_data)
    factories = []
    for key, contend in json_data.items():
        factory_get = get_value(contend, attr_name)
        # if factory_get != '':
        factories.append(factory_get)
    b = np.unique(factories)
    labels = {}
    label_list = []
    for item in b:
        num = factories.count(item)
        if num > baseline:
            labels[item] = num
            label_list.append(item)
    remains = {}
    print(len(json_data))
    filtered_data = {k:v for k,v in json_data.items() if attr_name in v}
    # 遍历主字典中的每个子字典
    for key, sub_dict in filtered_data.items():
        # 检查子字典的'type'值是否在类别数组中
        if sub_dict[attr_name] in label_list:
            # 如果是，则将该子字典添加到新的字典中
            remains[key] = sub_dict
        else:
            # 如果不是，则在新字典中为该键创建一个子字典，并将'type'设为""
            remains[key] = {attr_name: ''}
        # 打印新的字典以查看结果
    # print(len(remains))
    save_json('./labels', save_file, remains)
        

# pick_labels('factory', 50, f'./labels/check_result.json', f'/check_result_c1.json')
pick_labels('type', 90, f'./labels/check_result_c1.json', f'/check_result_c2.json')
# 1 manufactory 0 type
```
